```python
#import necessory libraries
import pandas as pd
```

```python
#load data
data=pd.read_csv(r"employe.csv")
```

```python
data.columns
```

| Column Name | Description |
|-------------|-------------|
| satisfactoryLevel | Employee satisfaction score ranging from 0 to 1 indicating how satisfied the employee is with their job. |
| lastEvaluation | The employee’s last performance evaluation score ranging from 0 to 1. |
| numberOfProjects | Total number of projects the employee has worked on in the company. |
| avgMonthlyHours | Average number of hours the employee works per month. |
| timeSpent.company | Number of years the employee has spent working at the company. |
| workAccident | Indicates whether the employee had a work accident (1 = Yes, 0 = No). |
| left | Target variable indicating whether the employee left the company (1 = Left, 0 = Stayed). |
| promotionInLast5years | Indicates whether the employee received a promotion in the last 5 years (1 = Yes, 0 = No). |
| dept | Department in which the employee works (e.g., Sales, HR, IT). |
| salary | Salary level of the employee categorized as low, medium, or high. |
