```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
#load data
data=pd.read_csv(r"final_employee_dataset(1).csv")
```

```python
#inputs and target 
y=data["left"]
x=data.drop("left",axis=1)
```

```python
#split data for training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
score=accuracy_score(model.predict(x_test),y_test)
score
```

```python
model.feature_importances_
```

```python
confusion_matrix=confusion_matrix(model.predict(x),y)
```

```python
import seaborn as sns
```

```python
sns.heatmap(confusion_matrix,annot=True)
```

- The model correctly predicts most employees who stay (10000).
- It also correctly identifies employees who leave (2000).
- Very few employees were incorrectly predicted as leaving (35).
- Only 2 employees who left were missed by the model.

The model performs extremely well with very few misclassifications, meaning it can reliably predict which employees are likely to leave the company.
