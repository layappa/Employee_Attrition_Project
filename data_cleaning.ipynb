## Data Cleaning Steps

1. **Import Dataset**  
   Load the dataset into the environment using libraries like Pandas.

2. **Create Dataset Copy**  
   Create a copy of the original dataset to avoid modifying the raw data.

3. **Check Missing Values**  
   Identify any missing values present in the dataset.

4. **Handle Missing Values**  
   Treat missing values using techniques such as removal or imputation.

5. **Check Duplicate Records**  
   Detect duplicate rows that may affect analysis.

6. **Fix Column Names**  
   Rename columns if they contain spaces, special characters, or inconsistent naming.

7. **Verify Data Types**  
   Ensure each column has the correct data type (numeric, categorical, etc.).

8. **Check Invalid Values**  
   Identify values outside the expected range or incorrect entries.

9. **Clean Categorical Variables**  
   Standardize category names and remove inconsistencies in categorical columns.

10. **Outlier Detection**  
    Detect extreme values that may distort analysis.

11. **Final Dataset Verification**  
    Perform a final check of the dataset structure, summary statistics, and sample rows.

```python
# import necessory libraries
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

1. **Import Dataset**

```python
data=pd.read_csv("employe.csv")
```

2. **Create Dataset Copy**

```python
df_copy=data.copy()
```

3. **Check Missing Values**

```python
df_copy.isnull().sum()
```

4. **Check Duplicate Records** 

```python
df_copy.duplicated().sum()
```

```python
# ratio of duplicate values
ratio=(df_copy.duplicated().sum())/(df_copy.shape[0])
print(ratio)
```

```python
# ratio of the duplicate rows is 20% so we are going drop all duplicate values
df_copy.drop_duplicates(inplace=True)
```

6. **Fix Column Names**

```python
df_copy.rename(columns={
    'satisfactoryLevel': 'satisfactory_Level',
    'lastEvaluation': 'last_Evaluation',
    'numberOfProjects': 'number_Of_Projects',
    'avgMonthlyHours': 'avg_Monthly_Hours',
    'timeSpent.company': 'time_Spent_Company',
    'workAccident': 'work_Accident',
    'promotionInLast5years': 'promotion_In_Last5years'
}, inplace=True)
```

```python
df_copy.columns
```

**Verify Data Types**

```python
df_copy.dtypes
```

**Check Invalid Values**

```python
for i in df_copy.columns:
    print(df_copy[i].value_counts())
```
